## 2. The Compass — No LLM needed

The core of the framework: evaluate any claim with three independent dimensions.

```python
from thirdanswer import Compass

# Coffee is good for your health
c = Compass(T=0.55, I=0.30, F=0.45)
print(f"Zone: {c.zone_emoji} {c.zone}")
print(f"Action: {c.zone_action}")
print(f"Confidence: {c.confidence:.2f}")
print(f"Paraconsistent: {c.is_paraconsistent} (T+F = {c.T + c.F:.2f})")
print(f"Should abstain: {c.should_abstain}")
```

```python
# Try different zones
examples = [
    ("MMR vaccine causes autism",        0.02, 0.03, 0.95),
    ("Remote work is more productive",   0.55, 0.25, 0.55),
    ("Quantum breaks encryption in 5yr", 0.15, 0.65, 0.30),
    ("Water boils at 100C at sea level", 0.95, 0.02, 0.01),
]

for claim, t, i, f in examples:
    c = Compass(T=t, I=i, F=f, label=claim)
    print(f"{c.zone_emoji} {c.zone:15s} | T={t:.2f} I={i:.2f} F={f:.2f} | {claim}")
```

## 4. Neutrosophic Statistics — What Classical Stats Can't Do

The key innovation: neutrosophic statistics can represent **paraconsistent** states (T + F > 1),
classify into **4 epistemic zones**, and **reduce indeterminacy** through algebraic cancellation
that interval statistics cannot achieve.

### 4.1 The Same P, Different Realities

4 drugs, all with classical P = 0.55. Classical stats says: "all moderately effective" (same action for all).
Neutrosophic stats classifies them into 4 DIFFERENT zones with 4 DIFFERENT actions.

```python
from thirdanswer import case_study_drug_efficacy

case_study_drug_efficacy()
```

### 4.2 Indeterminacy Reduction

Neutrosophic arithmetic preserves algebraic structure (variable I as symbol),
while interval arithmetic loses it. This means NS can *cancel* indeterminacy
that IS can only *grow*.

```python
from thirdanswer import NeutrosophicNumber, compare_uncertainty

# Smarandache's algebraic cancellation example
n1 = NeutrosophicNumber(4, 2)   # 4 + 2I
n2 = NeutrosophicNumber(4, -2)  # 4 - 2I

print("N1 = 4 + 2I, N2 = 4 - 2I")
print()

for op in ['add', 'mul', 'avg']:
    result = compare_uncertainty(n1, n2, op)
    print(f"{op.upper():5s}: NS uncertainty = {result['ns_uncertainty']:.1f}, "
          f"IS uncertainty = {result['is_uncertainty']:.1f}, "
          f"Winner: {result['winner']} "
          f"(ratio: {result['ratio']:.1f}x)")
```
